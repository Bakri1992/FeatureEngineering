```python
import pandas as pd
import numpy as np
```

```python
colspecs=[(0,11),(11,15),(15,17),(17,21)]
colspecs=colspecs+[(21 + 8*x, 21 + (8 * x)+5) for x in range(0, 31)]#
colspecs
```

```python
columns=["Station","Year","Month","Element"]
day_ranges=[i+1 for i in range(31)]
cols=columns+day_ranges
cols
```

```python

df=pd.read_fwf("weather.txt",colspecs=colspecs)
df.columns =cols
```

```python
df.head()
```

```python
df.replace(-9999,np.nan,inplace=True)
```

```python
df.head()
```

```python
df_melted=df.melt(id_vars=["Station","Year","Month","Element"]
           ,var_name="Day",value_name="Temperature")
```

```python
df_melted.head()
```

```python
df_pivoted=df_melted.pivot(index=["Station","Year","Month","Day"],
                    columns="Element",values="Temperature").reset_index()
```
