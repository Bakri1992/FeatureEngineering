```python
import pandas as pd
```

```python
start = 21
end = 266
step = 8

day_range = [(i, i+5) for i in range(start, end, step)]
print(day_range)
```

```python
colspecs=[(0,11),(11,15),(15,17),(17,21)]
colspecs=colspecs+day_range
colspecs
```

```python
df=pd.read_fwf("weather.txt",colspecs=colspecs)
```

```python
df_columns=["StationID","Year","Month","measure"]
day_columns=[(i+1) for i in range(31)]
df.columns=df_columns+day_columns
```

```python
df.columns
```

```python
df
```

```python
df1=pd.melt(df,id_vars=["StationID","Year","Month","measure"],var_name="Day",value_name="Temperatur")
```

```python
df1
```

```python
df1['Temperatur'] = pd.to_numeric(df1['Temperatur'], errors='coerce')
df2=df1.pivot_table(index= ['StationID', 'Year', 'Month', 'Day'],columns="measure",values="Temperatur").reset_index()
```
