```python
import requests
import pandas as pd
```

```python
base_url = 'https://services.odata.org/TripPinRESTierService/'
end_point = 'People'
```

```python
response = requests.get(
    url=base_url + end_point,
    params={
        '$format': 'json',
        '$top' : 30
    }
)
```

```python
data = response.json()
```

```python
data['value']
```

```python
pd.json_normalize(
    data['value'],
)
```

```python
pd.json_normalize(
    data['value'],
    "AddressInfo",
    max_level=1)
```

```python
df = pd.DataFrame(data['value'])
```
